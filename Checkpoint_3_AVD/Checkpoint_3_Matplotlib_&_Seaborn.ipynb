# **Visualisasi Data**

### Distribusi produk berdasarkan kategori

```python
plt.figure(figsize=(10,5))
sns.countplot(y=df['Category'], order=df['Category'].value_counts().index, palette='inferno')
plt.title("Distribusi Produk berdasarkan Kategori")
plt.xlabel("Jumlah Produk")
plt.ylabel("Kategori")
plt.show()
```

### Distribusi produk berdasarkan merek

```python
plt.figure(figsize=(10,5))
sns.countplot(y=df['Brand'], order=df['Brand'].value_counts().index, palette='Pastel1')
plt.title("Distribusi Produk berdasarkan Merek")
plt.xlabel("Jumlah Produk")
plt.ylabel("Merek")
plt.show()
```

### Distribusi harga produk

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Price'], bins=30, kde=True, color='lightcoral')
plt.title("Distribusi Harga Produk")
plt.xlabel("Harga")
plt.ylabel("Frekuensi")
plt.show()
```

### Distribusi rating produk

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Rating'], bins=30, kde=True, color='darksalmon')
plt.title("Distribusi Rating Produk")
plt.xlabel("Rating")
plt.ylabel("Frekuensi")
plt.show()
```

###Distribusi Produk Berdasarkan Warna

```python
plt.figure(figsize=(8, 5))
sns.countplot(x="Color", data=df, order=df["Color"].value_counts().index, palette="Set2")
plt.xticks(rotation=45)
plt.title("Distribusi Produk Berdasarkan Warna")
plt.xlabel("Warna")
plt.ylabel("Jumlah Produk")
plt.show()
```

### Distribusi Produk Berdasarkan Ukuran

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Size", order=df["Size"].value_counts().index, palette="pastel")
plt.title("Distribusi Produk Berdasarkan Ukuran")
plt.xlabel("Ukuran")
plt.ylabel("Jumlah Produk")
plt.show()
```

### Heatmap korelasi antar atribut numerik

```python
plt.figure(figsize=(8,6))
corr = df[['Price', 'Rating']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=1)
plt.title("Heatmap Korelasi Antar Atribut")
plt.show()
```

### Tren pembelian berdasarkan kategori

```python
df['Category'].value_counts().plot(kind='line', marker='o', figsize=(10,5), color='red')
plt.title("Tren Pembelian Berdasarkan Kategori")
plt.xlabel("Kategori")
plt.ylabel("Jumlah Pembelian")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

### Tren Produk Berdasarkan Ukuran

```python
plt.figure(figsize=(6,4))
sns.countplot(x=df['Size_Category'], order=df['Size_Category'].value_counts().index, palette='colorblind')
plt.title("Distribusi Produk Berdasarkan Ukuran")
plt.xlabel("Kategori ukuran")
plt.ylabel("Jumlah Produk")
plt.show()
```

### Tren Produk Berdasarkan Kategori Harga

### Distribusi Harga Berdasarkan Kategori Produk

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="Category", y="Price", palette="Set2")
plt.xticks(rotation=45)
plt.title("Distribusi Harga Berdasarkan Kategori Produk")
plt.xlabel("Kategori Produk")
plt.ylabel("Harga")
plt.show()
```

### Distribusi Rating Berdasarkan Kategori Produk
